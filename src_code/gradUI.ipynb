```python
import math
import librosa
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model
import gradio as gr
import re
from pydub import AudioSegment

SAMPLE_RATE = 22050
TRACK_DURATION = 30 
SAMPLES_PER_TRACK = SAMPLE_RATE * TRACK_DURATION
num_mfcc=13
n_fft=2048
hop_length=512
num_segments=10
samples_per_segment = int(SAMPLES_PER_TRACK / num_segments)
num_mfcc_vectors_per_segment = math.ceil(samples_per_segment / hop_length)

def trim_audio(file_path, max_duration=30000):
    audio = AudioSegment.from_file(file_path)
        duration = len(audio)
    
    if duration < max_duration:
        print("Sample is less than or equal to 30 seconds")
    else:
        audio = audio[:max_duration]
        audio.export(file_path+"_trimmed.wav", format="wav")
        print("Sample is more than or equal to 30 seconds. Trimmed to 30 seconds.")

model = tf.keras.models.load_model('genre_CLASS4.h5')
GENRE_MAPPING = [
  "Blues",
  "Classical",
  "Country",
  "Disco",
  "Hiphop",
  "Jazz",
  "Metal",
  "Pop",
  "Reggae",
  "rock"
]

def predict_genre(audio_file):  
    pattern = r".*\.wav$"  
    match = re.search(pattern, audio_file)
    match=match.string
    print(match)
    file_path = match
    trim_audio(match)
    file_path = match + "_trimmed.wav"
    signal, sample_rate = librosa.load(file_path, sr=SAMPLE_RATE)

    mfc=[]

    for d in range(num_segments):
        start = samples_per_segment * d
        finish = start + samples_per_segment
        mfcc = librosa.feature.mfcc(y=signal[start:finish],sr=sample_rate, n_mfcc=num_mfcc, n_fft=n_fft, hop_length=hop_length) 
        mfcc = mfcc.T
            if len(mfcc) == num_mfcc_vectors_per_segment:
            mfc.append(mfcc.tolist())
            mfcA=np.array(mfc)
            mfcA = mfcA[... ,np.newaxis]

    prediction = model.predict(mfcA)
    predicted_index = np.argmax(prediction, axis=1)[0]
    predicted_genre = GENRE_MAPPING[predicted_index]
    max_confidence = np.max(prediction[0])  
    confidence_percentage = max_confidence * 100
    return f"Predicted Genre: {predicted_genre}, Confidence: { confidence_percentage:.2f}%"

iface = gr.Interface(
    fn=predict_genre, 
    inputs=gr.Audio(type="filepath"),  
    outputs="text",  
    title="Music Genre Classifier")

iface.launch() 
```
